```python
import pandas as pd
# 데이터 프레임 만들기
df = pd.DataFrame(columns=['영화명', '개봉일', '대표국적 ', '장르 ', '감독 ', '배우 '])
# 엑셀 파일 불러와 데이터 프레임에 넣기
for i in range(1, 6):
    df_t = pd.read_excel(f"c:/data/movie_list/KOBIS_기간별박스오피스{i}.xlsx", engine="openpyxl", header=4)
    df_t.drop([0], inplace=True)
    df_t.drop([len(df_t.index)], inplace=True)
    df_t = df_t[['영화명', '개봉일', '대표국적 ', '장르 ', '감독 ', '배우 ']]
    df = pd.concat([df, df_t])
df
```

```python
df = df.dropna().reset_index(drop=True).reset_index().rename(columns={'index':'영화ID', '대표국적 ':'국적', '장르 ':'장르', '감독 ':'감독', '배우 ':'배우'})
df
```

```python
df.to_csv("../db/movie_list.csv")
```
